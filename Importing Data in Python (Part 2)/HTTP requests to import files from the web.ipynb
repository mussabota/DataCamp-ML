```python
### Get request using urllib
from urllib.request import urlopen, Request

url = 'https://www.wikipedia.org/'
request = Request(url)
response = urlopen(request)

html = response.read()
response.close()
html
```

```python
### Get request using  requests
import requests

url = 'https://www.wikipedia.org/'

r = requests.get(url)
text = r.text

text
```

```python
####  Scraping the web in Python
# HTML = Mix of unstructured and structured data
   # Structured data :
      # - Has pre - defined data model, or 
        # - Organized in a defined matter
```

```python
## BeautifulSoup 
# - Parse and extract structured data from HTML

from bs4 import BeautifulSoup
import requests

url = ''

r = requests.get(url)
html_doc = r.text

soup = BeautifulSoup(html_doc)

print(soup.prettify())

print(soup.title)

print(soup.get_text())
```

```python
## Exploring BeautifulSoup

for link in soup.find_all('a'):
    print(link.get('href'))
```
